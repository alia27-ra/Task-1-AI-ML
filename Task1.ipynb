```python
# ðŸ“Œ Step 1: Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# ðŸ“Œ Step 2: Load the Iris dataset using seaborn
iris = sns.load_dataset('iris')
```

```python
# ðŸ“Œ Step 3: Print the shape of the dataset
print("Shape of the dataset:", iris.shape)
```

```python
# ðŸ“Œ Step 4: Print column names
print("\nColumn Names:", iris.columns.tolist())
```

```python
# ðŸ“Œ Step 5: Print the first five rows using .head()
print("\nFirst 5 rows of the dataset:")
print(iris.head())
```

```python
# ðŸ“Œ Step 6: Show summary using .info()
print("\nDataset Info:")
print(iris.info())
```

```python
# ðŸ“Œ Step 7: Show descriptive statistics using .describe()
print("\nDescriptive Statistics:")
print(iris.describe())
```

```python
# ðŸ“Œ Step 8: Scatter plot (pairplot) to show relationships between features
sns.pairplot(iris, hue='species')
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
plt.show()
```

```python
# ðŸ“Œ Step 9: Histograms to show value distributions
iris.hist(figsize=(10, 6), color='lightgreen', edgecolor='black')
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

```python
# ðŸ“Œ Step 10: Box plots to identify outliers
plt.figure(figsize=(10, 6))
for i, column in enumerate(iris.columns[:-1], 1):  # skipping 'species'
    plt.subplot(2, 2, i)
    sns.boxplot(data=iris, x='species', y=column, palette="Set2")
    plt.title(f"Boxplot of {column} by Species")
plt.tight_layout()
plt.show()
```
